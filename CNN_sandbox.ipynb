```python
import torch
import torch.nn as nn
import torch.optim as optim
from GestureCNN.model import GestureCNN
from GestureCNN.dataset import train_loader, test_loader

model = GestureCNN(num_classes=10)
criterion = nn.CrossEntropyLoss()
optimizer = optim.Adam(model.parameters(), lr=0.001)
```

```python
epochs = 3
print("Starting Training Loop...\n")

for epoch in range(epochs):
    running_loss = 0.0
    correct = 0
    total = 0

    model.train()

    for batch_idx, (images, labels) in enumerate(train_loader):
        optimizer.zero_grad()
        outputs = model(images)
        loss = criterion(outputs, labels)
        loss.backward()
        optimizer.step()

        running_loss += loss.item()
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()

        if (batch_idx + 1) % 200 == 0:
            batch_accuracy = 100 * correct / total
            print(f"Epoch [{epoch+1}/{epochs}] | Batch [{batch_idx+1}/{len(train_loader)}] | Loss: {running_loss / 200:.4f} | Accuracy: {batch_accuracy:.2f}%")
            running_loss = 0.0

print("\n Training Complete")
```

```python
correct = 0
total = 0

model.eval()

with torch.no_grad():
    for images, labels in test_loader:
        outputs = model(images)
        _, predicted = torch.max(outputs.data, 1)
        total += labels.size(0)
        correct += (predicted == labels).sum().item()
final_test_accuracy = 100 * correct / total

print(f"→ Total Test Images Evaluated: {total}")
print(f"→ Correctly Classified: {correct}")
print(f"   Final Test Accuracy: {final_test_accuracy:.2f}%")
```
